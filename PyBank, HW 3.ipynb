```python
import os
import csv
```

```python
# Load file
budget_csv = os.path.join("..", "Resources", "budget_data.csv")
```

```python
# Read the csv and convert it into a list of dictionaries
with open(budget_csv) as budget:
    deposits = csv.reader(budget, delimiter=",")

    # Read the header
    header = next(deposits)

    # Track lists and counters
    total_months = 0
    net_total = 0
    avg_change = 0
    dates = []
    pl = []
    previous = []
    change = []
    greatest_increase = 0
    greatest_decrease = 0
    total_net = 0

    # Loop through data
    for row in deposits:
        # Manage data
        pl.append(int(row[1]))
        dates.append(row[0])
```

```python
    # Calculate the differences and adding them to the change list
    for value in range(1, len(pl)):
        change.append(int(pl[value] - int(pl[value-1])))
```

```python
    # Calculate total number of months
    total_months = len(dates)
```

```python
    # Calculate total profit and losses
    net_total = sum(pl)
```

```python
    # Calculate average of changes
    avg_change = sum(change) / len(change)
```

```python
    # Calculate greatest increase and decrease
    greatest_increase = max(change)
    greatest_decrease = min(change)
    incdex = change.index(greatest_increase)
    decdex = change.index(greatest_decrease)
    incdate = dates[incdex+1]
    decdate = dates[decdex+1]
```

```python
# Display results
print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {total_months}")
print(f"Total: ${net_total}")
print(f"Average Change: ${round(avg_change,2)}")
print(f"Greatest Increase in Profits: {incdate}, ${greatest_increase}")
print(f"Greatest Decrease in Profits: {decdate}, ${greatest_decrease}")
```

```python
# Export to a text file
text_file = open("budget_analysis.txt", "w")
text_file.write("Financial Analysis" + "\n")
text_file.write("----------------------------" + "\n")
text_file.write(f"Total Months: {total_months}" + "\n")
text_file.write(f"Total: ${net_total}" + "\n")
text_file.write(f"Average Change: ${round(avg_change,2)}" + "\n")
text_file.write(f"Greatest Increase in Profits: {incdate}, ${greatest_increase}" + "\n")
text_file.write(f"Greatest Decrease in Profits: {decdate}, ${greatest_decrease}")
text_file.close()
```
